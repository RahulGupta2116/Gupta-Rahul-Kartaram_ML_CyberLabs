```python
X_test=data_test.iloc[:,1:21].to_numpy()
X_test
```

```python
def knn_classification(X,y,X_test,k=3):
    prediction=[]
    for testpoint in X_test:
        distance=np.sqrt(np.sum((X-testpoint)**2,axis=1))
        k_indices=np.argsort(distance)[:k]
        k_value=[y[i] for i in k_indices]
        prediction.append(np.mean(k_value))
    return prediction
```

```python
prediction=knn_classification(X,y,X_test,k=3)
```

```python
prediction=pd.DataFrame(prediction)
pd.set_option('display.max_rows',None)
prediction 
```

```python
def f1_score(y,y_prediction):
    unique_classes=np.unique(y)
    f1_scores=[]
    total_samples=len(y)
    weights=[]
    for classes in unique_classes:
        tp=np.sum((y_prediction==classes)&(y==classes))
        fp=np.sum((y_prediction==classes)&(y!=classes))
        fn=np.sum((y_prediction!=classes)&(y==classes))
        precision=tp/(tp+fp)if(tp+fp)>0 else 0
        recall=tp/(tp+fn)if(tp+fn)>0 else 0
        f1=2*(precision*recall)/(precision+recall)if(precision+recall)>0 else 0
        f1_scores.append(f1)
        weights_f1=np.sum(np.sum(y==classes)/total_samples)
    weighted_f1=np.sum(np.array(f1_scores)*np.array(weights))
    return weighted_f1

        
```

```python
f1=f1_score(y,np.array(prediction))
print(f"f1_score_error:{f1}")
        
        
```
