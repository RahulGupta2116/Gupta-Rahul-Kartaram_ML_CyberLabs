```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Hours and Scores.csv")
```

```python
data
```

```python
X=data.Hours.to_numpy()
y=data.Scores.to_numpy()
```

```python
X,y
```

```python
plt.scatter(X,y,color="blue",label='Actual Data')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours vs Scores')
plt.show()
```

```python
total_error_values=[]
def loss_function(m,b,X,y):
    n=X.shape[0]
    error=0
    for i in range(n):
        error+=((m*X[i]+b)-y[i])**2
        total_error=(1/(2*n))*error
        return total_error
        
        
```

```python
def gradient_descent(m,b,L):
    n=X.shape[0]
    m_gradient=0
    b_gradient=0
    for i in range(n):
        m_gradient+=(1/n)*(((m*X[i]-b)-y[i])*X[i])
        b_gradient+=(1/n)*((m*X[i]-b)-y[i])
    m=m-m_gradient*L
    b=b-b_gradient*L
    total_error=loss_function(m,b,X,y)
    total_error_values.append(total_error)
    return m,b
    
   


        
        
```

```python
m=0
b=0
L=0.0001
iterations=1000
for i in range(iterations):
    if i % 50==0:
        print(f"iterations: {i}")
    m,b=gradient_descent(m,b,L)
   
     
```

```python
print(m,b)
```

```python
plt.scatter(X,y,color="blue",label='Actual Data')
#plt.plot(list(range(1,10)),[m*X+b for X in range(1,10)],color='red')
plt.plot(X,m*X+b,color='orange',label='Predicted Line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.title('Linear Regression Fit')
plt.show()
```

```python
 print("Total error values:")
for i, error in enumerate(total_error_values,start=1):
    print(f"iteration{i}:{error}")
```
